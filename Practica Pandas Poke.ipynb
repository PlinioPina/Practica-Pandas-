```python
#Mejor pokemon por type, usando mean para normalizar los datos

df = pd.read_csv("C:/Users/plini/Downloads/pokemon_data.csv")

df["MeanStats"] = df.iloc[:, 4:10].mean(axis=1)

best_by_mean = df.loc[df.groupby("Type 1")["MeanStats"].idxmax(),["Type 1", "Name"] + list(df.columns[4:10]) + ["MeanStats"]]

print(best_by_mean)
```

```python
#Top 5 por meanstats, usando mean para normalizar los datos

top5_mean = df.sort_values("MeanStats", ascending=False).head(5)

print(top5_mean[["Name", "Type 1"] + list(df.columns[4:10]) + ["MeanStats"]])
```

```python
#plot
import matplotlib.pyplot as plt

top10_mean = df.sort_values("MeanStats", ascending=False).head(10)
plt.figure(figsize=(10,6))

plt.bar(top10_mean["Name"], top10_mean["MeanStats"], color="skyblue", edgecolor="black")
plt.title("Top 10 Pokemon por promedio de stats", fontsize=14)
plt.xlabel("Pokemon", fontsize=12)
plt.ylabel("Promedio de Stats", fontsize=12)
plt.xticks(rotation=45)


plt.show()
```
